```python
from itertools import product
from collections import OrderedDict
from pyCBT.data.providers.mt.pairs import get_mt_pairs
from highcharts import Highchart, Highstock

import ffn

np = ffn.core.np
pd = ffn.data.pd
plt = ffn.core.plt
```

```python
pairs = get_mt_pairs()
pairs["EURUSD"]
```

```python
ohlc = pairs["EURUSD"].get("OPEN HIGH LOW CLOSE".split())
ohlc = np.column_stack((ohlc.index.values, ohlc.values))
volume = pairs["EURUSD"].get("VOLUME")
volume = np.column_stack((volume.index.values, volume.values))
groupingUnits = [
    [
        'week',
        [1]
    ],
    [
        'month',
        [1, 2, 3, 4, 6]
    ]
]

chart_options = {
    'rangeSelector': {
        'selected': 1
    },
    'title': {
        'text': 'EURUSD Historical'
    },

    'yAxis': [{
        'labels': {
            'align': 'right',
            'x': -3
        },
        'title': {
            'text': 'OHLC'
        },
        'height': '60%',
        'lineWidth': 2,
        'resize': {
            'enabled': True
        }
    }, {
        'labels': {
            'align': 'right',
            'x': -3
        },
        'title': {
            'text': 'Volume'
        },
        'top': '65%',
        'height': '35%',
        'offset': 0,
        'lineWidth': 2
    }],
    'tooltip': {
        'split': True
    }
}

chart = Highchart()
chart.set_dict_options(chart_options)
chart.add_data_set(ohlc, "candlestick", name="EURUSD", dataGrouping={"units": groupingUnits})
chart.add_data_set(volume, "column", name="Volume", yAxis=1, dataGrouping={"units": groupingUnits})
chart
```

```python
"""
Highstock Demos
Two panes, candlestick and volume: http://www.highcharts.com/stock/demo/candlestick-and-volume
"""
from highcharts import Highstock
from highcharts.highstock.highstock_helper import jsonp_loader
H = Highstock()

data_url = 'http://www.highcharts.com/samples/data/jsonp.php?filename=aapl-ohlcv.json&callback=?'
data = jsonp_loader(data_url, sub_d = r'(\/\*.*\*\/)')

ohlc = []
volume = []
groupingUnits = [
['week', [1]], 
['month', [1, 2, 3, 4, 6]]
]

for i in range(len(data)):
    ohlc.append(
        [
        data[i][0], # the date
        data[i][1], # open
        data[i][2], # high
        data[i][3], # low
        data[i][4]  # close
        ]
        )
    volume.append(
        [
        data[i][0], # the date
        data[i][5]  # the volume 
        ]
    )


options = {
    'rangeSelector': {
                'selected': 1
            },

    'title': {
        'text': 'AAPL Historical'
    },

    'yAxis': [{
        'labels': {
            'align': 'right',
            'x': -3
        },
        'title': {
            'text': 'OHLC'
        },
        'height': '60%',
        'lineWidth': 2
    }, {
        'labels': {
            'align': 'right',
            'x': -3
        },
        'title': {
            'text': 'Volume'
        },
        'top': '65%',
        'height': '35%',
        'offset': 0,
        'lineWidth': 2
    }],
}

H.add_data_set(ohlc, 'candlestick', 'AAPL', dataGrouping = {
                    'units': groupingUnits
                }
)
H.add_data_set(volume, 'column', 'Volume', yAxis = 1, dataGrouping = {
                    'units': groupingUnits
                }
)

H.set_dict_options(options)
```
